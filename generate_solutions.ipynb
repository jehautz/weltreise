```python
from itertools import permutations

cities = [
  {'name': 'Australien', 'x': 800, 'y': 595, 'order': 0},
  {'name': 'Sambia', 'x': 700, 'y': 242, 'order': 1},
  {'name': 'Kanada', 'x': 400, 'y': 325, 'order': 2},
  {'name': 'USA', 'x': 1247, 'y': 223, 'order': 3},
  {'name': 'Belgien', 'x': 492, 'y': 660, 'order': 4},
  {'name': 'China', 'x': 792, 'y': 382, 'order': 5},
  {'name': 'Taiwan', 'x': 1389, 'y': 665, 'order': 6},
  {'name': 'Brasilien', 'x': 198, 'y': 313, 'order': 7},
  {'name': 'Villach', 'x': 790, 'y': 250, 'order': 8}
]

orderings = permutations(cities)

costs = [# distance matrix w.r.t. city order as above
    [0, 109, 136, 148, 149, 74, 57, 160, 144],
    [109, 0, 141, 142, 76, 97, 109, 90, 70],
    [136, 141, 0, 25, 65, 89, 94, 93, 71],
    [148, 142, 25, 0, 75, 114, 118, 71, 82],
    [149, 76, 65, 75, 0, 77, 96, 88, 7],
    [74, 97, 89, 114, 77, 0, 20, 166, 73],
    [57, 109, 94, 118, 96, 20, 0, 184, 93],
    [160, 90, 93, 71, 88, 166, 184, 0, 92],
    [144, 70, 71, 82, 7, 73, 93, 92, 0]
]

preferences = [# preference list, entry [x, y] means city x before y
    ['Kanada', 'Sambia'],
    ['USA', 'Kanada'],
    ['Brasilien', 'Sambia'],
    ['Kanada', 'Taiwan'],
    ['USA', 'Brasilien'],
    ['Australien', 'USA'],
    ['USA', 'Australien'],
    ['Belgien', 'USA'],
    ['Taiwan', 'Sambia'],
    ['Kanada', 'Brasilien'],
    ['China', 'Belgien'],
    ['USA', 'China']
]
```

```python
def compute_costs(perm):
    ct = 0
    for i in range(len(perm)-1):
        ct += costs[perm[i]['order']][perm[i+1]['order']]
    return ct + costs[8][perm[0]['order']] + costs[perm[-1]['order']][8]

def compute_preferences(perm):
    ct = 0
    for p in preferences:
        for i1 in range(len(perm)):
            if perm[i1]['name'] == p[0]:
                break
        for i2 in range(len(perm)):
            if perm[i2]['name'] == p[1]:
                break
        if i1 < i2:
            ct += -400
    return ct

def construct_array(lst):
    lst = sorted(lst)
    values = set(lst)
    arr = []
    for val in values:
        arr.append([val, lst.index(val) + 1])
    return arr
```

```python
# mode 1: cheapest journey

results1 = []
orderings = permutations(cities[:-1])

for perm in orderings:
    results1.append(compute_costs(perm))
    
results1.sort()
```

```python
print(construct_array(results1))
```

```python
# mode 2: most preferences

results2 = []
orderings = permutations(cities[:-1])

for perm in orderings:
    results2.append(compute_preferences(perm))

results2.sort()
```

```python
print(construct_array(results2))
```

```python
# mode 3: mixed

results3 = []
orderings = permutations(cities[:-1])

for perm in orderings:
    results3.append(compute_costs(perm) + compute_preferences(perm))

results3.sort()
```

```python
print(construct_array(results3))
```
